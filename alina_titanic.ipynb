DATA ANALYSIS

```python
df.columns
```

```python
plt.figure(figsize=(15,6))
plt.ylabel('Age')
plt.plot(df['Age']   , color='green')
plt.show()
```

*The above plot shows the age of the passengers*

```python
#here 2 represents the death rate instead of 0
df.replace(0,2)
plt.ylabel('Survived')
plt.xlabel('Age')
plt.bar(df['Age'] , df['Survived'] )
plt.show()
```

*the above bar graph represents the survival rate according to age*

```python
plt.ylabel('Fare')
plt.xlabel('Gender')
plt.bar(df['Gender'] , df['Fare'] )
plt.show()
```

*Here we see that female passengers paid less compared to male passengers*

```python
plt.ylabel('Survived')
plt.xlabel('Gender')
plt.bar(df['Gender'] , df['Survived'] )
plt.show()
```

*The survival rate is same for both male and female*

```python
x = sns.countplot(x = 'Gender' , data = df)# using sea born
for bars in x.containers:
  x.bar_label(bars)
```

*there are 88 females and 95 males in all*

```python
x = sns.countplot(x = 'Survived' , data = df)# using sea born death rate is less bcs we dropped null rows
for bars in x.containers:
  x.bar_label(bars)
```

*Among the registered people 123 people survived and 60 were dead*

*From above two graphs we can conclude that from the registered people the no of males were more and survival rate is more that is 123 people survived among the registered people*

```python
sur_fare = df.groupby(['Survived'], as_index=False)['Fare'].sum().sort_values(by='Fare', ascending = False)
sns.barplot(data = sur_fare, x='Survived', y='Fare')
```

*Maximum who paid the fair were survived*
