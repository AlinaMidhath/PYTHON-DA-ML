```python
import matplotlib.pyplot as plt #py library used for plotting graphs most of thee req graphs are present in pyplot
```

```python
x = [1,2,3,4,5,6]# x and y shd have same dimensions (same no of elements)
y = [4,6,8,10,12,1]
```

```python
plt.xlabel('Days')
plt.ylabel('Stock')
plt.plot(x,y, marker= 'o' , linestyle = "-" , color = 'green', linewidth = 0.5) #to plot a graph, marker is used to show the points
plt.title('Sensex')
plt.show(x,y)# to show
```

```python
import numpy as np

x = np.array([1,7,4,5,9,3])
y = np.array([11,17,6,14,15,16,13,9])
z = np.array([1,2,3,5,6,7,10])

plt.plot(x, label="x")
plt.plot(y, label="y")
plt.plot(z, label="z")

plt.grid(axis = 'y')
plt.show
```

```python
import numpy as np
x = np.array([1,2,3,4,5,6,7])
y = x #eq of a st line
z = np.array([10,300,40,60,100,1000,120])
color = ['green']*(len(x)-1)
color.append('red')#having diff color for the last dot
plt.scatter(x,y , c = color , s=z, alpha= 0.75)#s=100 is to modify the size, and alpha is used tomodify the transperancy
plt.show()
```

Bar Graph

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(["A","B","C","D"])
y = np.array([10, 30, 12, 32])
#plt.bar(x,y , color ="teal" , width = 0.2)
plt.barh(x,y , color="blue", height = 0.2)
plt.show()
```

Histogram

```python
import numpy as np

x= np.random.normal(170,10,250)
#print(x)
plt.hist(x)
plt.show()
```
